```python
import random
import math
# Fish Count
# dalam 24 jam hitung ikan lewat setiap jam (random)
# 1. berapa jumlah ikan lewat yang terbanyak dan jam berapa
# 2. Menampilkan jumlah ekor ikan yang paling sedikit lewat dan jam berapa
# jam 1-22 ada mungkin sensor rusak yang ditandai dengan jumlah ikan -1 
# (jam kosong ganti dengan ikan pada ((jam+1) + (jam-1))/2 (floor))
def maximum(fish: list):
    max_number = fish[0]
    for i in range(len(fish)-1):
        if fish[i] > max_number:
            max_number = fish[i]
    return max_number
        
def minimum(fish: list):
    min_number = fish[0]
    for i in range(len(fish)-1):
        if fish[i] < min_number:
            min_number = fish[i]
    return min_number
        
jumlah_ikan = []
for i in range(24):
    if i == 0 or i == 23:
        jumlah_ikan.append(random.randint(0,10))
    else:
        if -1 not in jumlah_ikan:
            jumlah_ikan.append(random.randint(-1,10))
        else:
            jumlah_ikan.append(random.randint(0,10))

if -1 in jumlah_ikan:
    sensor_rusak = jumlah_ikan.index(-1)
jumlah_ikan[sensor_rusak] = math.floor((jumlah_ikan[sensor_rusak+1]+jumlah_ikan[sensor_rusak-1])/2)
ikan_terbanyak = maximum(jumlah_ikan)
ikan_sedikit = minimum(jumlah_ikan)
jam_ikan_terbanyak = jumlah_ikan.index(ikan_terbanyak)
jam_ikan_sedikit = jumlah_ikan.index(ikan_sedikit)

print(f"Ikan melewati sungai terbanyak pada jam {jam_ikan_terbanyak} dengan jumlah {ikan_terbanyak}")

print(f"Ikan melewati sungai tersedikit pada jam {jam_ikan_sedikit} dengan jumlah {ikan_sedikit}")
print(f"Sensor rusak pada pukul {sensor_rusak}")
```

```python
# CRUD (Create Read Update Delete)
# Ice cream store
# 2 buah list. List 1 jenis eskrim (minimal 3) dan list 2 harga
# Bikin function tambah rasa dan harga
# Update harga
# Hapus rasa dan harga 
eskrim = {
    "coklat" : 6000,
    "vanilla" : 7000,
    "mint" : 6500
}

def add_flavor():
    flavour = input("Input flavour name : ").lower()
    
    if eskrim.get(flavour) == None:
        while True:
            try:
                price = int(input("Enter a price : "))
                break
            except:
                print("Choose a number")
        eskrim[flavour] = price
        print(f"Added flavour { flavour} with a price of {price}")
    else:
        print("Flavour already exists")

def read_flavour():
    for i in eskrim:
        print(f"Flavour {i} with a price of {eskrim[i]}")
    
def update_price():
    print("Flavours available :")
    for i in eskrim:
        print(i.title())
    flavour = input("Choose a flavour to update : ").lower()
    if eskrim.get(flavour) == None:
        print("Flavour doesn't exist")
    else:
        while True:
            try:
                price = int(input("Enter a price : "))
                break
            except:
                print("Choose a number")

        eskrim[flavour] = price
        print(f"Updated price of {flavour} to {price}")

def delete_flavour():
    print("Flavours available :")
    for i in eskrim:
        print(i.title())
    flavour = input("Choose a flavour to delete : ").lower()
    if eskrim.get(flavour) == None:
        print("Flavour doesn't exist")
    else:
        del eskrim[flavour]
        print(f"Deleted {flavour.lower()} flavour")

while True:
    command = int(input(
"""
Choose a command :
1. Add flavour & price
2. Get flavour
3. Update price
4. Delete flavour & price
-1. Exit
"""
))
    if command in [1,2,3,4,-1]:
        if command == 1:
            add_flavor()
        if command == 2:
            read_flavour()
        if command == 3:
            update_price()
        if command == 4:
            delete_flavour()
        if command == -1:
            break
    else:
        print("Choose a valid command")
```

```python
def input_number():
    numbers = [int(x) for x in input("Input numbers (separated by space i.e : 1 2 3) : ").split()]
    return numbers
def sorting(lists, i = 0):
    if i == len(lists) - 1:
        print(lists) 
     
    max_index = i
    for j in range(i+1, len(lists)):
        if lists[j] < lists[max_index]:
            max_index = j
    lists[i], lists[max_index] = lists[max_index], lists[i]
    
    sorting(lists, i + 1)
numbers = input_number()
sorting(numbers)
```
